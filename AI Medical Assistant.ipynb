#Criteria :Question Answering using LLM


```python
- Load the large language model from Hugging Face
- Create a function to define the model parameters and generate a response


from transformers import pipeline

def model_param_resp_gen(question, context, model_name="deepset/roberta-base-squad2",max_seq_len=386, doc_stride=128, max_query_len=64,):

  question_pipeline = pipeline("question-answering", model=model_name)

  #define the params
  result = question_pipeline (
      question= question,
      context = context,
      max_seq_len=max_seq_len,
      doc_stride=doc_stride,
      max_query_len=max_query_len

  )
  return result
```

```python


Context ='''
The Merck Manuals are medical references published by the American pharmaceutical company Merck & Co., that cover a wide range of medical topics, including disorders, tests, diagnoses, and drugs. The manuals have been published since 1899 when Merck & Co. was still a subsidiary of the German company Merck.

Sepsis management in critical care requires early recognition, broad‑spectrum antibiotics within one hour, intravenous crystalloid fluids (30 mL/kg), vasopressors (norepinephrine) to maintain mean arterial pressure ≥65 mmHg, source control, and consideration of corticosteroids.

Appendicitis typically presents with periumbilical pain migrating to the right lower quadrant, anorexia, nausea, vomiting, and low‑grade fever. It cannot be cured with medication alone; the standard surgical procedure is appendectomy (open or laparoscopic).

Sudden patchy hair loss (alopecia areata) is often autoimmune. Possible causes include genetics, stress, or viral triggers. Effective treatments include intralesional corticosteroids, topical minoxidil, contact immunotherapy, and JAK inhibitors.

Traumatic brain injury (physical injury to brain tissue) management depends on severity. Mild cases require rest and monitoring; moderate‑severe cases need airway protection, intracranial pressure control (mannitol, hypertonic saline), seizure prophylaxis, surgical evacuation of hematomas, and neurorehabilitation.

Leg fracture during a hiking trip requires immediate immobilization with a splint, covering open wounds, applying ice, and evacuation. Definitive hospital treatment includes reduction, casting or surgical fixation (intramedullary nail, plates), followed by physical therapy and monitoring for complications.
'''

Questions = [
"What is the protocol for managing sepsis in a critical care unit?",
"What are the common symptoms of appendicitis, and can it be cured via medicine? If not, what surgical procedure should be followed to treat it?",
"What are the effective treatments or solutions for addressing sudden patchy hair loss, commonly seen as localized bald spots on the scalp, and what could be the possible causes behind it?",
"What treatments are recommended for a person who has sustained a physical injury to brain tissue, resulting in temporary or permanent impairment of brain function?",
"What are the necessary precautions and treatment steps for a person who has fractured their leg during a hiking trip, and what should be considered for their care and recovery?"
]
```

```python
- Apply the response generation function to get answers to the questions provided in the problem statement
for i in Questions:
  temp = model_param_resp_gen(i, Context)
  print(f"A.{i}: {temp} ")
```
